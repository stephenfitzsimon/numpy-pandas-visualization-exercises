```python
# 7. Write the code necessary to create a cross tabulation of the number of titles by department. 
# (Hint: this will involve a combination of SQL code to pull the necessary data and python/pandas 
# code to perform the manipulations.)

# I have a few answers to this question

sql = '''
SELECT dept_name, COUNT(DISTINCT title) AS num_of_titles
	FROM employees
		JOIN dept_emp USING (emp_no)
        JOIN titles USING (emp_no)
        JOIN departments USING (dept_no)
	GROUP BY dept_no;
'''

url = get_db_url('employees')

df_dept_titles = pd.read_sql(sql, url)
df_dept_titles
```

```python
#make it pretty looking?
df_dept_titles.set_index('dept_name').sort_values('dept_name')
```

```python
#pivot the table
df_dept_titles.pivot_table(columns='dept_name', values='num_of_titles')
```

```python
# alternatively, could find the number of employees with distinct titles in the depts.

#get a new df from sql database
#this will be more than just current employees
sql = '''
SELECT dept_name, title, count(emp_no)
	FROM employees
		JOIN dept_emp USING (emp_no)
        JOIN titles USING (emp_no)
        JOIN departments USING (dept_no)
	GROUP BY dept_no, title;
'''

url = get_db_url('employees')

df_dept_titles_alt = pd.read_sql(sql, url)
df_dept_titles_alt
```

```python
#make it pretty
df_dept_titles_alt.pivot_table(columns='dept_name', index='title', values='count(emp_no)')
```

```python
# just get the info and clean it in pandas
sql = '''
SELECT dept_name, title, emp_no
	FROM employees
		JOIN dept_emp USING (emp_no)
        JOIN titles USING (emp_no)
        JOIN departments USING (dept_no);
'''
url = get_db_url('employees')

df_dept_titles_info = pd.read_sql(sql, url)
df_dept_titles_info
```

```python
#if the question is asking how many unique titles are in each dept
df_dept_titles_info.groupby('dept_name').title.nunique()
```

```python
#if the question is asking the number of employees with that title in each dept
pd.crosstab(df_dept_titles_info.dept_name, df_dept_titles_info.title)
#this might make more sense than above because it has no NaN values; plus looks nicer because they are ints
```
