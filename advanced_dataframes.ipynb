```python
# 1. Copy the users and roles DataFrames from the examples above. 

users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['bob', 'joe', 'sally', 'adam', 'jane', 'mike'],
    'role_id': [1, 2, 3, 3, np.nan, np.nan]
})
users
```

```python
roles = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})
roles
```

```python
# 2. What is the result of using a right join on the DataFrames? 
#this depends on order of the join
users.merge(roles, how='right', left_on='role_id', right_on='id')
```

```python
roles.merge(users, how='right', right_on='role_id', left_on='id')
```

```python
# 3. What is the result of using an outer join on the DataFrames?
#both directions are the same, but in a different order
roles.merge(users, how='outer', left_on='id',  right_on='role_id')
```

```python
users.merge(roles, how='outer', left_on='role_id', right_on='id')
```

```python
# 4. What happens if you drop the foreign keys from the DataFrames and try to merge them?
users.drop(columns='role_id').merge(roles.drop(columns='id'), left_on='role_id', right_on='id')
#you won't be able to because there are not columns to merge on
```

```python
# 8. Check out your column names and perform any cleanup you may want on them.
#clarify some column names
mpg.rename(columns = {'cyls':'cylinders', 'hwy':'highway', 'cty':'city', 'trans':'transmission', 'displ':'displacement', 'drv':'drive'}, inplace=True)
mpg
```

```python
# 9. Display the summary statistics for the dataset.
mpg.describe().T
```

```python
# 10. How many different manufacturers are there?
mpg.manufacturer.unique().shape
```

```python
# 11. How many different models are there?
mpg.model.unique().shape
```

```python
# 12. Create a column named mileage_difference like you did in the DataFrames exercises; 
# this column should contain the difference between highway and city mileage for each car.
mpg['mileage_difference'] = mpg.highway - mpg.city
```

```python
# 13. Create a column named average_mileage like you did in the DataFrames exercises; this 
# is the mean of the city and highway mileage.
mpg['average_mileage'] = (mpg['city']+mpg['highway'])/2
```

```python
# 14. Create a new column on the mpg dataset named is_automatic that holds boolean values 
# denoting whether the car has an automatic transmission.
mpg['is_automatic'] = mpg.transmission.str.contains('auto')
```

```python
# 15. Using the mpg dataset, find out which which manufacturer has the best miles per gallon on average?
mpg.groupby('manufacturer').mean()['average_mileage'].nlargest(1, keep='all')
```

```python
# 16. Do automatic or manual cars have better miles per gallon?
mpg.groupby('is_automatic').mean()['average_mileage']
#manual cars have better mpg
```
