```python
# make the student grades data set
import pandas as pd
import numpy as np

np.random.seed(123)

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

# randomly generate scores for each student for each subject
# note that all the values need to have the same length here
math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))

df_gradebook = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})

type(df_gradebook)
```

```python
df_gradebook
```

```python
# a. Create a column named passing_english that indicates whether each student has a passing
# grade in english.

df_gradebook['passing_english'] = df_gradebook.english > 69
df_gradebook
```

```python
# b. Sort the english grades by the passing_english column. How are duplicates handled?
# looks like duplicates are sorted by the index value

df_gradebook.sort_values(by='passing_english')
```

```python
# c. Sort the english grades first by passing_english and then by student name. All the
# students that are failing english should be first, and within the students that are 
# failing english they should be ordered alphabetically. The same should be true for the 
# students passing english. (Hint: you can pass a list to the .sort_values method)

df_gradebook.sort_values(by=['passing_english', 'name'])
```

```python
# d. Sort the english grades first by passing_english, and then by the actual english 
# grade, similar to how we did in the last step.

df_gradebook.sort_values(by=['passing_english', 'english'])
```

```python
# can also pass list for ascending argument
df_gradebook.sort_values(by=['passing_english', 'english'], ascending=[False, True])

#but if sorted in the same way, just pass a boolean
df_gradebook.sort_values(by=['passing_english', 'english'], ascending=False)
```

```python
# e. Calculate each students overall grade and add it as a column on the dataframe. 
# The overall grade is the average of the math, english, and reading grades.
df_gradebook['overall_avg'] = sum([df_gradebook.math, df_gradebook.english, df_gradebook.reading])/3
df_gradebook
```

```python
# alternative.
# note you need to use the axis parameter to set it to row
df_gradebook['alt_avg'] = df_gradebook[['math', 'english', 'reading']].mean(axis=1)
df_gradebook
```

```python
# Rename the cty column to city.
mpg.rename(columns = {'cty':'city'}, inplace=True)
mpg
```

```python
# Rename the hwy column to highway.
mpg.rename(columns = {'hwy':'highway'}, inplace=True)
mpg
```

```python
# Do any cars have better city mileage than highway mileage?
# no returns
mpg[mpg.city > mpg.highway]
```

```python
# Create a column named mileage_difference this column should contain the difference 
# between highway and city mileage for each car.
mpg['mileage_difference'] = mpg.highway - mpg.city
mpg
```

```python
# Which car (or cars) has the highest mileage difference?
#recall: apply nlargest() needs a string for the column to pass
mpg.nlargest(1, 'mileage_difference', keep='all')
```

```python
# can also use a boolean mask
mpg[mpg.mileage_difference == mpg.mileage_difference.max()]
```

```python
# Which compact class car has the lowest highway mileage? The best?
# note that class is a reserved word, and therefore use the square bracket notation
mpg[mpg['class'] == 'compact'].nsmallest(1, 'highway', keep='all')
```

```python
# Create a column named average_mileage that is the mean of the city and highway mileage.
mpg['average_mileage'] = (mpg['city']+mpg['highway'])/2
mpg
```

```python
# Which dodge car has the best average mileage? The worst?
mpg[mpg['manufacturer'] == 'dodge'].nlargest(1, 'average_mileage', keep= 'all')
```

```python
mpg[mpg['manufacturer'] == 'dodge'].nsmallest(1, 'average_mileage', keep= 'all')
```

```python
# What is the overal percentage of specials?
# number of specials divided by the number of rows total

#number of special is given by:
mammals['specials'].value_counts()[True]
# = 10

#total number of rows from the shape tuple
mammals.shape[0]
# = 107

mammals['specials'].value_counts()[True]/mammals.shape[0]
```

```python
#alt: can sum booleans and divide
mammals.specials.sum()/len(mammals)
```

```python
#alt: can also just use mean
mammals.specials.mean()
```

```python
# How many animals are hoppers that are above the median speed? What percentage is this?

# get the median
mammals['speed'].median() # = 48

#get animals above the median that are also hoppers
mammals[(mammals['speed'] > mammals['speed'].median()) & (mammals['hoppers']==True)]

# get the number of rows for animals above medians
mammals[(mammals['speed'] > mammals['speed'].median()) & (mammals['hoppers']==True)].shape
#extract this into a variable
hoppers_above_median = mammals[(mammals['speed'] > mammals['speed'].median()) & (mammals['hoppers']==True)].shape[0]
print(hoppers_above_median)

#do the percentage calculation
hoppers_above_median/mammals.shape[0]
```
