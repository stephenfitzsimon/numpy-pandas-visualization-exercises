```python
import matplotlib.pyplot as plt
```

#### 1

Use matplotlib to plot the following equation:
y=x^2−x+2

You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.

```python
x = range(-5,7)
y = [i**2-i+2 for i in x]

plt.plot(x,y)
#zoom into the function minimum
#plt.xlim(-10,10)
#plt.ylim(-5,15)
#plot the origin
plt.text(0,0, '(0,0)')
plt.show()
```

#### 2
Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

y=√x

y=x^3

y=2^x

y=1/(x+1)

You can use functions from the math module to help implement some of the equations above.

```python
# square root function is only defined for non-negative reals 
x = range(1000)
y = [x**(0.5) for x in x]
plt.title('$\sqrt{x}$')
plt.plot(x,y)
plt.show()
```

```python
x = range(-100,100)
y = [x**3 for x in x]
plt.title('$x^3$')
plt.plot(x,y)
plt.show()
```

```python
#only need a low range to see how fast the increase is
#the negative x value will all be very close to 0
x = range(-1,10)
y = [2**x for x in x]
plt.title('$2^x$')
plt.plot(x,y)
plt.show()
```

```python
#avoid the asymptope at -1 and range does not take a float
x = range(0,20)
y = [1/(x+1) for x in x]
plt.title('$1/(x+1)$')
plt.plot(x,y)
plt.show()
```

#### 3
Combine the figures you created in the last step into one large figure with 4 subplots.

```python
#make the variables for the rows and columns
n_rows = 2
n_cols = 2
plt.figure(figsize = (10,10))

#subplot 1
plt.subplot(n_rows, n_cols, 1)
x = range(1000)
y = [x**(0.5) for x in x]
plt.plot(x, y)
plt.title('$\sqrt{x}$')

#subplot 2
plt.subplot(n_rows,n_cols,2)
x = range(-100,100)
y = [x**3 for x in x]
plt.plot(x, y)
plt.title('$x^3$')

#subplot 3
plt.subplot(n_rows,n_cols,3)
x = range(-1,10)
y = [2**x for x in x]
plt.plot(x,y)
plt.title('$2^x$')

#subplot 4
plt.subplot(n_rows,n_cols,4)
x = range(0,20)
y = [1/(x+1) for x in x]
plt.plot(x,y)
plt.title('1/(x+1)')

plt.show()
```

```python

plt.figure(figsize=(10,10))
x = list(range(20))
y = [x**(0.5) for x in x]
plt.plot(x, y, c='green',label='$\sqrt{x}$')


x = list(range(-10,10))
y = [x**3 for x in x]
plt.plot(x, y, c='red',label='$x^3$')

x = list(range(-1,12))
y = [2**x for x in x]
plt.plot(x,y, c='blue',label='$2^x$')

x = list(range(0,11))
y = [1/(x+1) for x in x]
plt.plot(x,y,c='orange',label='$1/(x+1)$')

plt.xlim(-1,10)
plt.ylim(-1,10)
plt.legend()
```
