#### big O notation



```python
import matplotlib.pyplot as plt
import math as m
import numpy as np
```

```python


plt.figure(figsize=(15,15))
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.xlim(1,15)
plt.ylim(0,10)

x = range(1,20)
#one_x = [x for x in range(1,20)]
one_y = [0*x+1 for x in one_x]
plt.plot(x, one_y, c='red', label='$O(1)$')

log_x = [x for x in range(1,20)]
log_y = [m.log(x) for x in log_x]
plt.plot(x, log_y, c='blue', label='$O(\log{n})$')

n_x = [x for x in range(1,20)]
n_y = [x for x in n_x]
plt.plot(x, n_y, c='green', label='$O(n)$')

nlogn_x = x
nlogn_y = [n*m.log(n) for n in x]
plt.plot(nlogn_x, nlogn_y, c='grey', label='$O(n\log{n})$')

sq_x = [x for x in range(1, 10)]
sq_y = [x**2 for x in sq_x]
plt.plot(sq_x, sq_y, c='orange', label='$O(n^2)$')

exp_x = [x for x in range(1,7)]
exp_y = [2**x for x in exp_x]
plt.plot(exp_x, exp_y, c='purple', label='$O(2^x)$')

fact_x = [x for x in range(1,6)]
fact_y = [m.factorial(x) for x in fact_x]
plt.plot(fact_x, fact_y, c = 'black', label='$O(n!)$')

nn_x = [x for x in range(1,5)]
nn_y = [x**x for x in nn_x]
plt.plot(nn_x, nn_y, c = 'cyan', label='$O(n^n)$')

plt.legend()
```

```python
#some of these make more sense as a real valued funcitons
plt.figure(figsize=(15,15))

x = np.linspace(1,100)
y = np.log10(x)

plt.plot(x,y,'r')
plt.show()
```
