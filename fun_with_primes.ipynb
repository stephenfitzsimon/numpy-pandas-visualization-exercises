```python
import matplotlib.pyplot as plt
```

```python
#get a list of primes up to n

def is_prime(n):
    #returns true if n is prime
    if n<=1:
        return False
    else:
        for i in range(2, int(n**(0.5))+1):
            if n%i==0:
                return False
        return True

def get_primes(n):
    #return a list of primes up to n
    primes = []
    for x in range(2,n):
        if is_prime(x):
            primes.append(x)
    return primes
            
primes = get_primes(1000000)
```

```python
#plot the primes
n = 1000
x = [x for x in range(n)]
y = [x if x in primes else None for x in range(n)]

plt.figure(figsize=(10,10))
plt.scatter(x,y)
```

```python
#get the distance between two consecutive primes
differences = [primes[x]-primes[x-1] for x in range(1,len(primes))]
```

```python
#what is the distribution of the differences between primes?
plt.figure(figsize=(10,10))

bin_n = [x for x in range(1,max(differences))]
plt.hist(differences, bins = bin_n)
plt.title('distribution of differences between consecutive primes')
max(differences)
```

```python
#plot only the maximum distance between primes for primes less than x

def get_max_difference(n):
    #get the maximum distance between primes less than n
    p_less = [p for p in primes if p < n]
    differences = [p_less[x]-p_less[x-1] for x in range(1,len(p_less))]
    try:
        max_diff = max(differences)
        return max_diff
    except ValueError:
        return 0

x = [x for x in range(1,10000)]
y = [get_max_difference(x) for x in x]
print(type(max))

plt.figure(figsize=(10,10))
plt.scatter(x,y)
```
